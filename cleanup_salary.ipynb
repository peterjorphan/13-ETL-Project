```python
import pandas as pd
import os
```

```python
path = 'Resources\salary_raw.csv'
```

```python
# read csv file as DataFrame and display
df = pd.read_csv(path, index_col=0)
df.head()
```

```python
# Clean rows
df = df[df.RK != 'RK']
df.head()
```

```python
# Split name from position and create new column
s = df['Name'].str.split(', ', expand=True)
df['Name'] = s[0]
df['Position'] = s[1]
df.head()
```

```python
# Reorder columns
df = df[['Season', 'RK', 'Name', 'Position', 'Team', 'Salary']]
df.head()
```

```python
# Save to csv
df.to_csv('Resources\salary.csv')
```
